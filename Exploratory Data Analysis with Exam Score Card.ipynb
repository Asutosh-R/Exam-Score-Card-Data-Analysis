```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# Checking Null Values:

df.isnull().sum()
```

```python
val= df.gender.value_counts()

plt.figure(figsize=[10,6])
plt.title("Percentace of Male and Female", fontsize=15)
plt.pie(data=df, x=val, autopct="%1.0f%%", labels=["Female","Male"])
plt.show()
```

### Observation 
* Number of female are more than number of male in Exam Scores
* There is 52% Female and 48% Male in the Exam Scores

```python
# Plotting Graph according Race/Ethnicity with gender

plt.figure(figsize=[10,6])
sns.countplot(data=df, x="race/ethnicity", hue="gender")
```

```python
# Plotting Graph According to the Parent's level of Education with male and Female
plt.figure(figsize=[10,6])
sns.countplot(data=df, x="parental level of education", hue="gender").set_xlabel("Parental Level of Education", fontsize = 15)
```

```python
plt.figure(figsize=[15,5])

plt.subplot(1,2,1)
sns.countplot(data=df, x="lunch", hue="gender")

plt.subplot(1,2,2)
sns.countplot(data=df, x="test preparation course", hue="gender")


plt.show()
```

```python
plt.figure(figsize=[20,5])

plt.subplot(1,3,1)
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
sns.boxplot(data=df, x="gender", y="math score")

plt.subplot(1,3,2)
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
sns.boxplot(data=df, x="gender", y="reading score")

plt.subplot(1,3,3)
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
sns.boxplot(data=df, x="gender", y="writing score")
```

### Observation

* Math Score - Males have scored maximum score than Female.
* Females have scored better in both reading and writing

```python
plt.figure(figsize=[20,5])

plt.subplot(1,3,1)
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
plt.xticks(rotation=90)
sns.boxplot(data=df, x="parental level of education", y="math score")

plt.subplot(1,3,2)
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
plt.xticks(rotation=90)
sns.boxplot(data=df, x="parental level of education", y="reading score")

plt.subplot(1,3,3)
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
plt.xticks(rotation=90)
sns.boxplot(data=df, x="parental level of education", y="writing score")

plt.show()
```

### Observation
* Children whose parents have master degree has performed better
